```python
#Q5.

"""
'url_for()' function is used for URL building. It generates a URL for the specified endpoint, which is defined in the routes of our Flask application. The url_for() function takes the name of the endpoint as its first argument and any additional keyword arguments representing the variables used in the route pattern.
"""

from flask import Flask, url_for

app = Flask(__name__)

@app.route("/")
def home():
    return "<h1>Welcome to the Home Page</h1>"

@app.route("/about")
def about():
    return "<h1>This is the About Page</h1>"

@app.route("/contact")
def contact():
    return "<h1>Contact us at xyz@example.com</h1>"

@app.route("/user/<username>")
def user_profile(username):
    return f"Hello, {username}!"


if __name__=="__main__":
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
        print(url_for('user_profile', username = 'Zoey'))
```
